```python
missing_columns = informatii.columns[informatii.isnull().any()]

for column in missing_columns:
    num_missing = informatii[column].isnull().sum()
    print(f'Coloana "{column}":')
    print(f'Numrul de valori lips: {num_missing}')

    total_entries = len(informatii[column])
    proportion_missing = (num_missing / total_entries) * 100
    print(f'Procentul de valori lips: {proportion_missing:.2f}%')
    
    print('-' * 50)
```

```python
bins = [0, 20, 40, 60, informatii['Age'].max()]
labels = ['0-20', '21-40', '41-60', '61+']

informatii['AgeCategory'] = pd.cut(informatii['Age'], bins=bins, labels=labels, right=False)

age_category_counts = informatii['AgeCategory'].value_counts().sort_index()

print(age_category_counts)

plt.figure(figsize=(10, 6))
sns.countplot(data=informatii, x='AgeCategory', order=labels)
plt.title('Distribuia pasagerilor pe categorii de vrst')
plt.xlabel('Categorie de vrst')
plt.ylabel('Numr de pasageri')
plt.grid(True)
plt.show()
```

```python
survivors_by_age_category = informatii[(informatii['Sex'] == 'male') & (informatii['Survived'] == 1)].groupby('AgeCategory').size()

total_by_age_category = informatii[informatii['Sex'] == 'male'].groupby('AgeCategory').size()

survival_rate_by_age_category = (survivors_by_age_category / total_by_age_category) * 100

plt.figure(figsize=(10, 6))
sns.barplot(x=survival_rate_by_age_category.index, y=survival_rate_by_age_category.values)
plt.title('Procentul de supravieuire al brbailor pe categorii de vrst')
plt.xlabel('Categorie de vrst')
plt.ylabel('Procentul de supravieuire (%)')
plt.grid(True)
plt.show()
```

```python
copii = informatii[informatii['Age'] < 18]
adulti = informatii[informatii['Age'] >= 18]

procent_copii = (len(copii) / len(informatii)) * 100
print(f"Procentul copiilor aflai la bord: {procent_copii:.2f}%")

rata_supravietuire_copii = (copii['Survived'].sum() / len(copii)) * 100
rata_supravietuire_adulti = (adulti['Survived'].sum() / len(adulti)) * 100

supravietuire_df = pd.DataFrame({
    'Categorie': ['Copii', 'Aduli'],
    'Rata de supravieuire (%)': [rata_supravietuire_copii, rata_supravietuire_adulti]
})

plt.figure(figsize=(10, 6))
sns.barplot(x='Categorie', y='Rata de supravieuire (%)', data=supravietuire_df)
plt.title('Rata de supravieuire pentru copii i aduli')
plt.xlabel('Categorie')
plt.ylabel('Rata de supravieuire (%)')
plt.grid(True)
plt.show()
```

```python
informatii = pd.read_csv('train.csv')
# Calculm media vrstelor rotunjite inferioare pentru fiecare clas
class_ages = informatii.groupby('Pclass')['Age'].mean().apply(np.floor)

# Funcie pentru completarea valorilor lips n funcie de clas i rotunjirea inferioar
def complete_age(row):
    if pd.isnull(row['Age']):
        return class_ages[row['Pclass']]
    else:
        return row['Age']

# Aplicm funcia pentru a completa valorile lips n coloana 'Age'
informatii['Age'] = informatii.apply(complete_age, axis=1)

# Completm valorile lips din coloanele categoriale cu cea mai frecvent valoare pentru fiecare clas
for column in informatii.select_dtypes(include='object').columns:
    mode_class = informatii.groupby('Pclass')[column].apply(lambda x: x.mode()[0])
    informatii[column].fillna(informatii['Pclass'].map(mode_class), inplace=True)

informatii.to_csv('informatii_modificate.csv', index=False)
```

```python

informatii['Title'] = informatii['Name'].str.extract(' ([A-Za-z]+)\.')

informatii['Verificare'] = (informatii['Title'].isin(['Mr', 'Mrs', 'Miss', 'Ms', 'Master'])) & (informatii['Sex'] == 'male') | (informatii['Title'].isin(['Mrs', 'Miss', 'Ms'])) & (informatii['Sex'] == 'female')


numar_persoane = informatii.groupby('Title')['Verificare'].sum()

plt.figure(figsize=(10, 6))
numar_persoane.plot(kind='bar', color='skyblue')
plt.title('Numrul de persoane pentru fiecare titlu de nobilime')
plt.xlabel('Titlu de nobilime')
plt.ylabel('Numrul de persoane')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
